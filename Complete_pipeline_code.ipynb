```python
import cv2
def apply_clahe_np(image_np):
    """
    image_np: (H, W, 3) float32 numpy array in [0,1]
    returns enhanced float32 array in [0,1]
    """
    img = (image_np * 255).astype(np.uint8)

    # Extract green channel (best for vessels)
    g = img[:, :, 1]

    # Apply CLAHE to green channel
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    g_clahe = clahe.apply(g)

    # Replace green channel
    img[:, :, 1] = g_clahe

    # Convert back to float32
    return img.astype(np.float32) / 255.0

@tf.function
def apply_clahe_tf(image):
    """
    Wraps apply_clahe_np so it can be used in tf.data pipelines.
    """
    enhanced = tf.numpy_function(apply_clahe_np, [image], tf.float32)
    enhanced.set_shape(image.shape)  # important!
    return enhanced
```

```python
ROI_IN_VESSEL_SIZE = 512

# ==================================================
# STEP 3: EXTRACT VESSELS INSIDE ROI
# ==================================================
def extract_vessels_in_roi(vessel_map, roi_mask):
    """Extract only vessel pixels inside the ROI region."""
    # Resize ROI mask to match vessel map
    roi_mask_resized = cv2.resize(roi_mask, (ROI_IN_VESSEL_SIZE, ROI_IN_VESSEL_SIZE))

    # Normalize ROI mask to 0/1
    roi_mask_resized = (roi_mask_resized > 0).astype(np.uint8)

    # Expand to 3 channels for RGB multiplication
    roi_mask_3ch = np.repeat(roi_mask_resized[:, :, np.newaxis], 3, axis=2)

    # Mask the vessel map â€” keep only within ROI
    vessels_in_roi = vessel_map * roi_mask_3ch

    return vessels_in_roi


# ==================================================
# STEP 4: VISUALIZATION
# ==================================================
def visualize_results(original_image, vessel_map, vessels_in_roi):
    plt.figure(figsize=(15, 6))
    
    plt.subplot(1, 3, 1)
    plt.title("Original Fundus Image")
    plt.imshow(original_image)
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.title("Full Vessel Map")
    plt.imshow(vessel_map)
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.title("Vessels Inside ROI Only")
    plt.imshow(vessels_in_roi)
    plt.axis("off")

    plt.tight_layout()
    plt.show()


# ==================================================
# STEP 5: RUN
# ==================================================
vessels_in_roi = extract_vessels_in_roi(vessel_map, roi_mask)
visualize_results(original_image, vessel_map, vessels_in_roi)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.morphology import skeletonize, remove_small_objects
from scipy.ndimage import distance_transform_edt, label

def compute_vessel_diameters_per_segment(vessels_in_roi, min_area=40):
    """
    Compute average diameter for each individual vessel segment (artery & vein).
    vessels_in_roi: RGB vessel map (Red=Artery, Blue=Vein)
    min_area: minimum pixel area to keep a vessel component
    """
    all_diameters = {}     # to store all per-vessel diameters
    labeled_components = {}

    # --- Separate Artery and Vein Masks ---
    artery_mask = vessels_in_roi[:, :, 0] > 0.5
    vein_mask   = vessels_in_roi[:, :, 2] > 0.5

    for label_name, mask in zip(['Artery', 'Vein'], [artery_mask, vein_mask]):
        print(f"\nProcessing {label_name}...")
        if np.sum(mask) == 0:
            all_diameters[label_name] = []
            continue

        # --- Remove small fragments ---
        mask_cleaned = remove_small_objects(mask, min_size=min_area)
        labeled_mask, num_components = label(mask_cleaned)

        per_vessel_diams = []

        # --- Compute diameter for each labeled component ---
        for i in range(1, num_components + 1):
            comp_mask = labeled_mask == i
            if np.sum(comp_mask) < min_area:
                continue

            # Skeletonize this vessel
            skel = skeletonize(comp_mask)

            # Distance transform gives radius at each pixel
            dist = distance_transform_edt(comp_mask)
            local_diams = 2 * dist[skel]

            if len(local_diams) > 0:
                mean_d = np.mean(local_diams)
                per_vessel_diams.append(mean_d)

        all_diameters[label_name] = per_vessel_diams
        labeled_components[label_name] = labeled_mask

        # --- Print summary ---
        if len(per_vessel_diams) > 0:
            print(f"  Found {len(per_vessel_diams)} {label_name.lower()} segments.")
            for j, d in enumerate(per_vessel_diams, 1):
                print(f"    Segment {j}: {d:.2f} px")
            print(f"  â†’ Mean {label_name} diameter: {np.mean(per_vessel_diams):.2f} px")
        else:
            print(f"  No valid {label_name.lower()} segments detected.")

    return all_diameters, labeled_components
```

```python
import numpy as np

def combine_pair_rms(w1, w2):
    """Combine two widths with RMS rule (Knudtson-style pairwise)."""
    return np.sqrt((w1**2 + w2**2) / 2.0)

def reduce_to_one_equivalent(widths):
    """
    Reduce a list of widths to a single equivalent by pairwise RMS combining.
    Strategy: sort widths descending and combine the two largest repeatedly.
    This is a common practical strategy; order has minor effect for similar widths.
    """
    widths = list(widths)  # copy
    if len(widths) == 0:
        return None
    if len(widths) == 1:
        return float(widths[0])

    # Work on floats
    widths = [float(w) for w in widths]
    # combine until one remains
    while len(widths) > 1:
        widths.sort(reverse=True)           # combine largest two first (common practice)
        a = widths.pop(0)
        b = widths.pop(0)
        c = combine_pair_rms(a, b)
        widths.append(c)
    return widths[0]

def compute_crae_crve_avr(diameters_dict, n_big=6, debug=False):
    """
    diameters_dict: {'Artery':[d1,d2,...], 'Vein':[d1,d2,...]}
    n_big: number of largest vessels to use (default 6 = 'big six')
    returns: dict with CRAE, CRVE, AVR and the lists used
    """
    # fetch lists (may be empty)
    arteries = sorted(diameters_dict.get('Artery', []), reverse=True)[:n_big]
    veins    = sorted(diameters_dict.get('Vein',    []), reverse=True)[:n_big]

    if debug:
        print("Arteries used:", arteries)
        print("Veins used:", veins)

    art_equiv = reduce_to_one_equivalent(arteries)
    ven_equiv = reduce_to_one_equivalent(veins)

    if art_equiv is None or ven_equiv is None:
        # not enough data to compute both CRAE and CRVE
        return {
            'CRAE': None, 'CRVE': None, 'AVR': None,
            'arteries_used': arteries, 'veins_used': veins
        }

    CRAE = 0.88 * art_equiv
    CRVE = 0.95 * ven_equiv
    AVR  = CRAE / CRVE if CRVE != 0 else None

    return {
        'CRAE': float(CRAE),
        'CRVE': float(CRVE),
        'AVR': float(AVR) if AVR is not None else None,
        'arteries_used': arteries,
        'veins_used': veins,
        'arteriole_equivalent': float(art_equiv),
        'venule_equivalent': float(ven_equiv)
    }
```

```python
import os
import cv2
import numpy as np
import pandas as pd
import tensorflow as tf
from skimage.morphology import skeletonize, remove_small_objects
from scipy.ndimage import distance_transform_edt, label

# ==================================================
# CONFIG
# ==================================================
IMAGE_SIZE_VESSEL = 512
IMAGE_SIZE_OD = 256

input_dir = "/kaggle/working/severity_4_images"   
output_csv = "/kaggle/working/avr4_results.csv"

# Your trained models
vessel_model = loaded_model          
od_model = model                    


# ==================================================
# STEP 1: PREPROCESS + PREDICT FUNCTIONS
# ==================================================
def preprocess_image(image_path, size):
    """Load and preprocess image for prediction."""
    image = tf.io.read_file(image_path)
    image = tf.io.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [size, size])
    image = tf.cast(image, tf.float32) / 255.0
    image = apply_clahe_tf(image)
    return tf.expand_dims(image, axis=0)


def predict_vessel_map(image_path):
    """Predict artery/vein vessel map."""
    img = preprocess_image(image_path, IMAGE_SIZE_VESSEL)
    pred_mask = vessel_model.predict(img, verbose=0)
    pred_mask = (pred_mask > 0.5).astype("float32")

    artery = pred_mask[0, :, :, 0]
    vein = pred_mask[0, :, :, 1]

    vessel_map = np.zeros((IMAGE_SIZE_VESSEL, IMAGE_SIZE_VESSEL, 3), dtype=np.float32)
    vessel_map[:, :, 0] = artery
    vessel_map[:, :, 2] = vein
    return vessel_map


def predict_optic_disc(image_path):
    """Predict optic disc mask using OD model (trained on 256Ã—256)."""
    img = preprocess_image(image_path, IMAGE_SIZE_OD)
    pred_mask = od_model.predict(img, verbose=0)
    pred_mask = (pred_mask > 0.5).astype("uint8")
    mask = np.squeeze(pred_mask)
    return mask


# ==================================================
# STEP 2: ROI COMPUTATION
# ==================================================
def get_roi_mask(image_path, od_mask):
    """Compute annular ROI mask between 0.5D and 1D from OD edge."""
    original = cv2.imread(image_path)
    original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
    original = cv2.resize(original, (od_mask.shape[1], od_mask.shape[0]))

    contours, _ = cv2.findContours(od_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(contours) == 0:
        return None

    c = max(contours, key=cv2.contourArea)
    (x, y), radius = cv2.minEnclosingCircle(c)
    center = (int(x), int(y))

    r_inner = 2.0 * radius
    r_outer = 3.0 * radius

    h, w = od_mask.shape
    roi_mask = np.zeros((h, w), dtype=np.uint8)
    cv2.circle(roi_mask, center, int(r_outer), 255, -1)
    cv2.circle(roi_mask, center, int(r_inner), 0, -1)

    # resize ROI mask to match vessel map (512Ã—512)
    roi_mask_resized = cv2.resize(roi_mask, (IMAGE_SIZE_VESSEL, IMAGE_SIZE_VESSEL))
    roi_mask_resized = (roi_mask_resized > 0).astype(np.uint8)
    return roi_mask_resized


# ==================================================
# STEP 3: EXTRACT VESSELS IN ROI
# ==================================================
def extract_vessels_in_roi(vessel_map, roi_mask):
    roi_mask_3ch = np.repeat(roi_mask[:, :, np.newaxis], 3, axis=2)
    vessels_in_roi = vessel_map * roi_mask_3ch
    return vessels_in_roi


# ==================================================
# STEP 4: DIAMETER COMPUTATION
# ==================================================
def compute_vessel_diameters_per_segment(vessels_in_roi, min_area=40):
    """
    Compute average diameter for each vessel segment.
    Returns per-segment diameters for Artery and Vein.
    """
    all_diameters = {}

    artery_mask = vessels_in_roi[:, :, 0] > 0.5
    vein_mask = vessels_in_roi[:, :, 2] > 0.5

    for label_name, mask in zip(['Artery', 'Vein'], [artery_mask, vein_mask]):
        if np.sum(mask) == 0:
            all_diameters[label_name] = []
            continue

        mask_cleaned = remove_small_objects(mask, min_size=min_area)
        labeled_mask, num_components = label(mask_cleaned)

        per_vessel_diams = []
        for i in range(1, num_components + 1):
            comp_mask = labeled_mask == i
            if np.sum(comp_mask) < min_area:
                continue
            skel = skeletonize(comp_mask)
            dist = distance_transform_edt(comp_mask)
            local_diams = 2 * dist[skel]
            if len(local_diams) > 0:
                per_vessel_diams.append(np.mean(local_diams))

        all_diameters[label_name] = per_vessel_diams

    return all_diameters


# ==================================================
# STEP 5: CRAE/CRVE/AVR CALCULATION (Knudtson)
# ==================================================
def combine_pair_rms(w1, w2):
    """Combine two widths with RMS rule (Knudtson-style pairwise)."""
    return np.sqrt((w1**2 + w2**2) / 2.0)


def reduce_to_one_equivalent(widths):
    """Reduce a list of widths to a single equivalent by pairwise RMS combining."""
    widths = list(widths)
    if len(widths) == 0:
        return None
    if len(widths) == 1:
        return float(widths[0])
    widths = [float(w) for w in widths]
    while len(widths) > 1:
        widths.sort(reverse=True)
        a = widths.pop(0)
        b = widths.pop(0)
        c = combine_pair_rms(a, b)
        widths.append(c)
    return widths[0]


def compute_crae_crve_avr(diameters_dict, n_big=6, debug=False):
    arteries = sorted(diameters_dict.get('Artery', []), reverse=True)[:n_big]
    veins = sorted(diameters_dict.get('Vein', []), reverse=True)[:n_big]

    if debug:
        print("Arteries used:", arteries)
        print("Veins used:", veins)

    if len(arteries) < 3 or len(veins) < 3:
        if debug:
            print("âš ï¸ Not enough segments for reliable CRAE/CRVE computation.")
        return {'CRAE': None, 'CRVE': None, 'AVR': None,
                'arteries_used': arteries, 'veins_used': veins}

    art_equiv = reduce_to_one_equivalent(arteries)
    ven_equiv = reduce_to_one_equivalent(veins)

    CRAE = 0.88 * art_equiv
    CRVE = 0.95 * ven_equiv
    AVR = CRAE / CRVE if CRVE != 0 else None

    return {
        'CRAE': float(CRAE),
        'CRVE': float(CRVE),
        'AVR': float(AVR) if AVR is not None else None,
        'arteries_used': arteries,
        'veins_used': veins,
        'arteriole_equivalent': float(art_equiv),
        'venule_equivalent': float(ven_equiv)
    }


# ==================================================
# STEP 6: MAIN LOOP â€” PROCESS ALL IMAGES
# ==================================================
results = []

for img_name in os.listdir(input_dir):
    if not img_name.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    img_path = os.path.join(input_dir, img_name)
    print(f"Processing {img_name} ...")

    #  Predict optic disc
    od_mask = predict_optic_disc(img_path)
    roi_mask = get_roi_mask(img_path, od_mask)
    if roi_mask is None:
        print("âš ï¸ No OD detected â€” skipping.")
        continue

    #  Predict vessels
    vessel_map = predict_vessel_map(img_path)

    #  Extract ROI vessels
    vessels_in_roi = extract_vessels_in_roi(vessel_map, roi_mask)

    #  Compute per-segment diameters
    diameters = compute_vessel_diameters_per_segment(vessels_in_roi)

    #  Compute CRAE, CRVE, AVR
    avr_results = compute_crae_crve_avr(diameters, n_big=6, debug=False)

    results.append({
        "image_name": img_name,
        "CRAE": avr_results['CRAE'],
        "CRVE": avr_results['CRVE'],
        "AVR": avr_results['AVR'],
        "Arteries_Used": avr_results['arteries_used'],
        "Veins_Used": avr_results['veins_used']
    })


# ==================================================
# STEP 7: SAVE RESULTS
# ==================================================
df_results = pd.DataFrame(results)
df_results.to_csv(output_csv, index=False)

print("\nâœ… AVR computation complete (Knudtson method)")
print(f"Results saved to: {output_csv}")
df_results.head()
```

```python
import pandas as pd

# ==================== Load and Preprocess ====================
df = pd.read_csv('/kaggle/input/vgg16-avr-results/avr1_results (3).csv')
print(f"Total images in CSV: {len(df)}")

# Keep only valid AVR values (< 0.95 to remove outliers or errors)
df = df[df['AVR'].notnull() & (df['AVR'] < 1.0)].copy()
valid_images = len(df)
print(f"Number of images with valid AVR < 1.0: {valid_images}")

# Extract patient ID (assuming image_name like "1000_left.jpeg" or "1000_L_1.png")
df['patient_id'] = df['image_name'].apply(lambda x: str(x).split('_')[0])

# ==================== Round AVR and Define Risk Categorization ====================
df['AVR_rounded'] = df['AVR'].round(2)

def assess_cvd_risk(avr_value):
    if avr_value >= 0.75:
        return "Low Risk"
    elif 0.67 <= avr_value < 0.75:
        return "Moderate Risk"
    else:
        return "High Risk"

df['Risk_Level'] = df['AVR_rounded'].apply(assess_cvd_risk)

# ==================== Select Each Patientâ€™s Highest Risk ====================
# Define a numeric order so we can select the highest-risk level per patient
risk_order = ['Low Risk', 'Moderate Risk', 'High Risk']
risk_rank = {r: i for i, r in enumerate(risk_order)} 

# Assign numeric rank and select max rank per patient
df['risk_rank'] = df['Risk_Level'].map(risk_rank)

# Select the row corresponding to the highest risk (max rank) for each patient
patient_highest_risk = (
    df.loc[df.groupby('patient_id')['risk_rank'].idxmax()]
    .reset_index(drop=True)
)

print(f"ðŸ‘¥ Unique patients after selecting highest risk: {len(patient_highest_risk)}")

# ==================== Count Patients per Risk Category ====================
summary = (
    patient_highest_risk['Risk_Level']
    .value_counts()
    .reindex(risk_order)
    .reset_index()
    .rename(columns={'index': 'Risk_Level', 'Risk_Level': 'Unique_Patient_Count'})
    .fillna(0)
)

# ==================== Display Results ====================
print("\nðŸ“Š Summary of AVR-Based CVD Risk Levels (One per Patient, Worst Eye Considered):")
print(summary)
```
